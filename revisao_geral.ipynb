## **Operações básicas**

```python
titulo = 'CALCULADORA'

print('================ {:^20} ================'.format(titulo))
print()

print('1 - ADIÇÃO (+)')
print('2 - SUBTRAÇÃO (-)')
print('3 - MULTIPLICAÇÃO (*)')
print('4 - DIVISÃO (/)')
print('5 - Resto da divisão (MOD %)')

opcao = int(input('Digite a opção desejada: '))
print()

a = int(input('Digite um número: '))
b = int(input('Digite outro número: '))
print()

if opcao == 1:
  soma = a + b
  print('Resposta -> A soma entre {} e {} vale {}.'.format(a, b, soma))
elif opcao == 2:
  sub = a - b
  print('Resposta -> A subtração entre {} e {} vale {}.'.format(a, b, sub))
elif opcao == 3:
  multi = a * b
  print('Resposta -> A multiplicação entre {} e {} vale {}.'.format(a, b, multi))
elif opcao == 4:
  div = a / b
  print('Resposta -> A divisão entre {} e {} vale {}.'.format(a, b, div))
elif opcao == 5:
  mod = a % b
  print('Resposta -> O resto da divisão entre {} e {} vale {}.'.format(a, b, mod))
```

1 - Operações e Variáveis
Crie um programa que imprima (print) os principais indicadores da loja Hashtag&Drink no último ano.
Obs: faça tudo usando variáveis.

Valores do último ano:

Quantidade de Vendas de Coca = 150<br>
Quantidade de Vendas de Pepsi = 130<br>
Preço Unitário da Coca = 1,50 <br>
Preço Unitário da Pepsi = 1,50<br>
Custo da Loja: 2.500,00<br>

1.1 Qual foi o faturamento de Pepsi da Loja?

1.2. Qual foi o faturamento de Coca da Loja?

1.3. Qual foi o Lucro da loja?

1.4. Qual foi a Margem da Loja? (Lembre-se, margem = Lucro / Faturamento). Não precisa formatar em percentual

```python
titulo = 'Hashtag&Drink'

print('=.=.=.=.=.=.=.=.=.=.=.=.{:^10}=.=.=.=.=.=.=.=.=.=.=.=.'.format(titulo))

qtd_coca = 150
qtd_pepsi = 130
unit_coca = 1.50
unit_pepsi = 1.50
custo_loja = 2500
faturamento = (qtd_coca * unit_coca) + (qtd_pepsi * unit_pepsi)
lucro = ((qtd_coca * unit_coca) + (qtd_pepsi * unit_pepsi)) - custo_loja
margem = lucro / faturamento
print()

print('O faturamento de Pepsi da loja foi de R${:.2f}'.format(qtd_pepsi * unit_pepsi).replace('.',','))
print('O faturamento de Coca da loja foi de R${:.2f}'.format(qtd_coca * unit_coca).replace('.',','))

if lucro < 0:
  print('O prejuizo da loja foi de R${:+.2f}'.format(lucro).replace('.',','))
else:
  print('O lucro da loja foi de R${:+.2f}'.format(lucro).replace('.',','))

print('A margem da loja foi de R${:+.2f}'.format(margem))
```

1 - Cálculo de Bônus

Crie um programa que calcule e dê um print no bônus que os funcionários devem receber segundo a regra:

A meta é 1000 vendas.<br> 
Se o valor de vendas for maior ou igual a meta, o valor do bônus do funcionário é 10% do valor de vendas.<br>
Caso contrário o valor de bônus do funcionário é 0.<br>
Print o bônus dos 3 funcionários


vendas_funcionario1 = 1000
vendas_funcionario2 = 770
vendas_funcionario3 = 2700


```python
titulo = 'CALCULO DE BÔNUS'
print('====================== {:^30} ======================'.format(titulo))
print()

meta = 1000
vendas_f1 = 2500
vendas_f2 = 770
vendas_f3 = 2700

if vendas_f1 >= meta:
  if vendas_f1 >= 2000:
    bonus1 = 0.15 * vendas_f1
    print('-> funcionario 1: Ganhou R$ {:.2f} de bônus, totalizando R$ {:.2f}'.format(bonus1, vendas_f1 + bonus1).replace('.',','))
  else:
    bonus1 = 0.10 * vendas_f1
    print('-> funcionario 1: Ganhou R$ {:.2f} de bônus, totalizando R$ {:.2f}'.format(bonus1, vendas_f1 + bonus1).replace('.',','))
else:
  print('-> funcionario 1: Não ganhou bônus esse ano, faltou {} para atingir a pontuação minima.'.format(meta - vendas_f1))

if vendas_f2 >= meta:
  if vendas_f2 >= 2000:
    bonus2 = 0.15 * vendas_f2
    print('-> funcionario 2: Ganhou R$ {:.2f} de bônus, totalizando R$ {:.2f}'.format(bonus2, vendas_f2 + bonus2).replace('.', ','))
  else:
    bonus1 = 0.10 * vendas_f1
    print('-> funcionario 2: Ganhou R$ {:.2f} de bônus, totalizando R$ {:.2f}'.format(bonus1, vendas_f1 + bonus1).replace('.',','))
else:
  print('-> funcionario 2: Não ganhou bônus esse ano, faltou {} para atingir a pontuação minima.'.format(meta - vendas_f2))

if vendas_f3 >= meta:
  if vendas_f3 >= 2000:
    bonus3 = 0.15 * vendas_f3
    print('-> funcionario 3: Ganhou R$ {:.2f} de bônus, totalizando R$ {:.2f}'.format(bonus3, vendas_f3 + bonus3).replace('.', ','))
  else:
    bonus1 = 0.10 * vendas_f3
    print('-> funcionario 3: Ganhou R$ {:.2f} de bônus, totalizando R$ {:.2f}'.format(bonus1, vendas_f1 + bonus1).replace('.',','))
else:
  print('-> funcionario 3: Não ganhou bônus esse ano, faltou {} para atingir a pontuação minima.'.format(meta - vendas_f3))
```

2 - Criando um mini sistema de controle de estoque

- Crie um sistema para ser usado pelo time de controle de estoque de um centro de distribuição.
- Imagine que ao fim de todo dia, o time conta quantas unidades de produto existem no estoque. Se tivermos um estoque abaixo do estoque permitido para aquela categoria do produto, o time deve ser avisado (print) para fazer um novo pedido daquele produto.
- Cada categoria de produto tem um estoque mínimo diferente, segundo a regra abaixo:

- alimentos -> Estoque mínimo: 50
- bebidas -> Estoque mínimo: 75
- limpeza -> Estoque mínimo: 30

Para isso vamos criar um programa que pede 3 inputs do usuário: nome do produto, categoria e quantidade atual em estoque.

Se o produto tiver abaixo do estoque mínimo da categoria dele, o programa deve printar a mensagem "Solicitar {produto} à equipe de compras, temos apenas {unidades} em estoque"

Exemplo: Se o usuário preenche os inputs com: bebidas, dolly, 90, o programa não deve exibir nenhuma mensagem.<br>
Agora, se o usuário preenche os inputs com: bebidas, guaraná, 60, o programa deve exibir a mensagem "Solicitar guaraná à equipe de compras, temos apenas 60 unidades em estoque.

Obs: lembre de usar o int() para transformar o número inserido pelo usuário no input de string para int.<br>
Obs2: Caso o usuário não preencha alguma das 3 informações, o programa deve exibir uma mensagem para avisá-lo de preencher corretamente.

```python
print('================= {} ================='.format('CONTROLE DE ESTOQUE'))
print()

produto = input('Nome do produto: ')
categoria = input('Categoria do produto: ')
qtd = input('Quantidade atual do produto: ')
print()

if produto and categoria and qtd:
  qtd = int(qtd)
  if categoria == 'alimentos' and qtd < 50:
    print('Solicitar: SIM')
    print('Quantidade em estoque: {} unidades'.format(qtd))
    print('Quantidade minima para ser solicitada: {}'.format(50 - qtd))
  elif categoria == 'bebidas' and qtd < 75:
    print('Solicitar: SIM')
    print('Quantidade em estoque: {} unidades'.format(qtd))
    print('Quantidade minima para ser solicitada: {}'.format(75 - qtd))
  elif categoria == 'limpeza' and qtd < 30:
    print('Solicitar: SIM')
    print('Quantidade em estoque: {} unidades'.format(qtd))
    print('Quantidade minima para ser solicitada: {}'.format(30 - qtd))
else:
  print('Preencha os dados corretamente')
  
```

#### EXERCICIO

Crie um programa para fazer uma consulta de estoque. O usuário do programa deve inserir o nome do produto e, caso ele não exisa na lista, ele será avisado. Caso exista, o programa deve dizer a quantidade de unidades em estoque do produto.

```python
lista = ['cebola', 'tomate', 'batata']
estoque = [1500, 2500, 150]
produto = input('Informe o nome do produto: ')
i = lista.index(produto)

produto = produto.strip()

if produto in lista:
  print('{} esta na lista'.format(produto))
  qtd_estoque = estoque[i]
  print(qtd_estoque)
else:
  print('Não tem no estoque')
```
